```python
import pandas as pd
import random
lst = ['robot'] * 10
lst += ['human'] * 10
random.shuffle(lst)
data = pd.DataFrame({'whoAmI':lst})
data.head()
```

```python
pd.get_dummies(data['whoAmI'])
```

```python
def human(x):
  if x == 'human':
    return '1'
  elif x == 'robot':
    return '0'
def robot(x):
  if x == 'robot':
    return '1'
  elif x == 'human':
    return '0'
data['human'] = data.whoAmI.apply(human)
data['robot'] = data.whoAmI.apply(robot)
data[['human', 'robot']]
```
